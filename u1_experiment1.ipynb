```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.datasets import make_blobs

# Implement vector operations
def dot_product(v1, v2):
    return np.dot(v1, v2)

def cross_product(v1, v2):
    return np.cross(v1, v2)

def vector_norm(v):
    return np.linalg.norm(v)

# vectors
v1 = np.array([2, 3])
v2 = np.array([4, -1])

print("Dot Product:", dot_product(v1, v2))
print("Cross Product:", cross_product(v1, v2))
print("Norm of v1:", vector_norm(v1))

# Generate a 2D dataset
X, y = make_blobs(n_samples=100, centers=2, random_state=42, cluster_std=1.5)

# Analyze linear separability

plt.figure(figsize=(6,6))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm', edgecolors='k')
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Generated Dataset")
plt.grid(True)
plt.show()

# Task 4: Train SVM and visualize decision boundary

clf = SVC(kernel='linear')
clf.fit(X, y) 

# Plot decision boundary
xx, yy = np.meshgrid(np.linspace(X[:,0].min()-1, X[:,0].max()+1, 100),
                     np.linspace(X[:,1].min()-1, X[:,1].max()+1, 100))
Z = clf.decision_function(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.figure(figsize=(6,6))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm', edgecolors='k')
plt.contour(xx, yy, Z, levels=[0], colors='black', linestyles='dashed')
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("SVM Classification Boundary")
plt.grid(True)
plt.show()

# visualization: Norms of feature vectors
norms = np.linalg.norm(X, axis=1)
plt.figure(figsize=(6,6))
plt.scatter(X[:, 0], X[:, 1], c=norms, cmap='viridis', edgecolors='k')
plt.colorbar(label='Vector Norm')
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Norm of Feature Vectors")
plt.grid(True)
plt.show()

# visualization: Dot product similarity between points

dot_products = np.dot(X, v1)  # Projection onto vector v1
plt.figure(figsize=(6,6))
plt.scatter(X[:, 0], X[:, 1], c=dot_products, cmap='plasma', edgecolors='k')
plt.colorbar(label='Dot Product with v1')
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Dot Product Similarity")
plt.grid(True)
plt.show()
```
